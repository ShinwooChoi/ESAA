# 1. Lightgbm

* 데이터 '와인품질분류.zip'입니다.

1. 필요한 라이브러리를 가져옵니다.

```python
import pandas as pd
import lightgbm as lgbm
```

2. 데이터를 불러옵니다.

```python
train = pd.read_csv('/content/train.csv')
test = pd.read_csv('/content/test.csv')
smpl_sub = pd.read_csv('/content/sample_submission (1).csv')
```

```python
train.head()
```

```python
test.head()
```

3. 변수를 변환합니다.

```python
# type에는 white와 red 두 종류가 있습니다.
# 각각 0,1로 변환합니다.
# 'type' 변수: white → 0, red → 1 로 변환
train['type'] = train['type'].map({'white': 0, 'red': 1})
test['type'] = test['type'].map({'white': 0, 'red': 1})
```

```python
train.head()
```

```python
test.head()
```

```python
# 모델에 입력하기 전 데이터를 정형합니다.
# train_x : train의 'index', 'quality' 피처를 제거하고 저장합니다.
# train_y : train의 'quality'를 저장합니다.
# test_x : test의 'index'를 제거하고 저장합니다.

# train 데이터를 학습용 X, y로 나눔
train_x = train.drop(columns=['index', 'quality'])
train_y = train['quality']

# test 데이터를 예측용 X로 나눔
test_x = test.drop(columns=['index'])
```

```python
train_x.shape, train_y.shape, test_x.shape
```

```python
train_x.head()
```

```python
train_y.head()
```

```python
test_x.head()
```

4. LightGBM 모델을 생성 및 훈련합니다.

```python
import lightgbm as lgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

# train 데이터를 학습/검증 세트로 분리
X_train, X_test, y_train, y_test = train_test_split(train_x, train_y, test_size=0.2, random_state=42)

# 모델 생성
model = lgb.LGBMRegressor(random_state=42)

# 모델 훈련
model.fit(X_train, y_train)

# 검증 예측 및 성능 확인
y_pred = model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
print('검증 mse:', mse)
```

5. 생성한 모델로 예측 데이터를 생성합니다.

```python
# 예측 결과 생성
test_pred = model.predict(test_x)
```

6. 제출파일을 생성합니다.

```python
smpl_sub.head()
```

```python
#서브미션 파일의 quality 변수에 예측값을 대입합니다.
# sample submission 파일에 예측 결과 대입
smpl_sub['quality'] = test_pred
```

```python
smpl_sub.head()
```

```python
# csv 파일로 저장합니다.
smpl_sub.to_csv('submission.csv', index=False)
```

# 2. XGBoost

```python
# 데이터 다운로드 링크로 데이터를 코랩에 불러옵니다.

!wget 'https://bit.ly/3i4n1QB'

import zipfile
with zipfile.ZipFile('3i4n1QB', 'r') as existing_zip:
    existing_zip.extractall('data')
```

```python
# 라이브러리 및 데이터 불러오기
import pandas as pd
from xgboost import XGBClassifier

train = pd.read_csv('/content/train.csv')
test = pd.read_csv('/content/test.csv')

train.drop('index',axis = 1 ,inplace =True)
test.drop('index',axis = 1 ,inplace =True)
```

```python
# 데이터 확인
train.head()
```

```python
test.head()
```

```python
# 원핫 인코딩 (pd.get_dummies())
train = pd.get_dummies(train, columns=['type'])
test = pd.get_dummies(test, columns=['type'])
```

```python
# 모델 정의
from xgboost import XGBRegressor
model = XGBRegressor(random_state=42)
```

```python
# 모델 학습
# X 는 train에서 quality 를 제외한 모든 변수
# y 는 train의 qulity 변수
X = train.drop(['quality'],axis=1)
y = train['quality']

# fit 메소드를 이용해 모델 학습
model.fit(X, y)
```

```python
# predict 메소드와 원핫 인코딩한 test 데이터를 이용해 품질 예측
test_pred = model.predict(test)
```

```python
# sample_submission.csv 파일을 불러와 예측된 값으로 채워 주기
submission = pd.read_csv('/content/sample_submission (1).csv')
submission['quality'] = test_pred
```

```python
# 정답파일 내보내기
submission.to_csv('xgb_pred.csv',index = False)
```

# 3. 보팅

```python
# 데이터 다운로드 링크로 데이터를 코랩에 불러옵니다.

!wget 'https://bit.ly/3i4n1QB'

import zipfile
with zipfile.ZipFile('3i4n1QB', 'r') as existing_zip:
    existing_zip.extractall('data')
```

```python
# 라이브러리 및 데이터 불러오기
import pandas as pd
from lightgbm import LGBMClassifier
from xgboost import XGBClassifier

from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import VotingClassifier

train = pd.read_csv('data/train.csv')
test = pd.read_csv('data/test.csv')

train.drop('index',axis = 1 ,inplace =True)
test.drop('index',axis = 1 ,inplace =True)
```

```python
# 데이터 확인
train.head()
```

```python
test.head()
```

```python
# 원핫 인코딩 (pd.get_dummies())
train = pd.get_dummies(train)
test = pd.get_dummies(test)

test = test.reindex(columns=train.columns.drop('quality'), fill_value=0)
```

```python
# 모델 정의 (LightGBM, XGBoost, Random Forest)
from lightgbm import LGBMClassifier
from xgboost import XGBClassifier
from sklearn.ensemble import RandomForestClassifier, VotingClassifier

lgbm = LGBMClassifier(random_state=42)
xgb = XGBClassifier(random_state=42)
rf = RandomForestClassifier(random_state=42)

# VotingClassifier 정의
voting_clf = VotingClassifier(
    estimators=[('lgbm',lgbm), ('xgb', xgb), ('rf', rf)],
    voting='hard')
```

```python
# 모델 학습

# X 는 train에서 quality 를 제외한 모든 변수
# y 는 train의 qulity 변수
X = train.drop(['quality'], axis=1)
y = train['quality']

# fit 메소드를 이용해 모델 학습
voting_clf.fit(X,y)
```

```python
# predict 메소드와 원핫 인코딩한 테스트 데이터를 이용해 품질 예측
test_pred = voting_clf.predict(test)
```
